```python
def getdate():
    import datetime
    return datetime.datetime.now()


def Retrieve():
    while(True):
        
        print("Please select the mode:")
        print("Press 1 : Diet")
        print("Press 2 : Exercise")
        print("Press 3 : Exit")

        userInput = input("Enter your choice:")
        if userInput == '1':
            print("Diet Mode")
            print("\n")
            print("Select the user")
            print("\n")
            print("Press 1 : Harry")
            print("Press 2 : Rohan")
            print("Press 3 : Hammad")
        
        if userInput == '3':
            print("Thank You....")
            break

            uInput = input("Enter your user:")
            if uInput == '1':
                print("\n")
                f = open("Harry_Diet.txt","r")
                print(f.read())
                print("\n")
                f.close()

            elif uInput == '2':
                print("\n")
                f = open("Rohan_Diet.txt","r")
                print(f.read())
                print("\n")
                f.close()

            elif uInput == '3':
                print("\n")
                f = open("Hammad_Diet.txt","r")
                print(f.read())
                print("\n")
                f.close()

        if userInput == '2':
            print("Exercise Mode")
            print("\n")
            print("Select the user.")
            print("Press 1 : Harry")
            print("Press 2 : Rohan")
            print("Press 3 : Hammad")

            uInput = input("Enter your user:")
            if uInput == '1':
                print("\n")
                f = open("Harry_Exercise.txt","r")
                print(f.read())
                print("\n")
                f.close()

            elif uInput == '2':
                print("\n")
                f = open("Rohan_Exercise.txt","r")
                print(f.read())
                print("\n")
                f.close()

            elif uInput == '3':
                print("\n")
                f = open("Hammad_Exercise.txt","r")
                print(f.read())
                print("\n")
                f.close()
            
        continue
                

def manageSystem():
    print("\t\t\t\t Welcome to our Health Management System:)")
    while(True):
        print("Please select your mode: ")
        print("Press 1 : Log")
        print("Press 2 : Retrieve")
        print("Press 3:  Exit")

        print("\n")
        user_inputOne = int(input(f"Enter your choice:"))

        if user_inputOne == 1:
            print("You have selected Log mode.")
            print("\n")
            print("Where you want to log?")
            print("1.Harry")
            print("2.Rohan")
            print("3.Hammad")

            user_inputTwo = int(input(f"Please select the user:"))
            print("\n")

            if user_inputTwo == 1:
                print("Welcome Harry")
                print("Where you want to enter?")
                print("1. Diet ")
                print("2. Execerise")

                user_inputThree = input(f"Enter your choice:")
                if user_inputThree == '1':
                    print("Diet Section")

                    while(True):
                        userMeal = input("Enter the Meal you have taken:")
                        if userMeal == 'q' or userMeal == 'Q':
                            print("Successfully inserted")
                            break
                        elif userMeal != 'q' or userMeal != 'Q':
                            f = open("Harry_Diet.txt","a")
                            f.write(f"\n [{getdate()}]: {userMeal}")
                            f.close()
                            continue

                elif user_inputThree == '2':
                    print("Excerise Section")

                    while(True):
                        userExercise = input("Enter the Exercise that you have Done:")
                        if userExercise == 'q' or userExercise == 'Q':
                            print("Successfully inserted")
                            break
                        elif userExercise != 'q' or userExercise != 'Q':
                            f = open("Harry_Exercise.txt","a")
                            f.write(f"\n [{getdate()}]: {userExercise}")
                            f.close()
                            continue

                elif user_inputThree != '1' and user_inputThree != '2':
                    print("Wrong Input")


            elif user_inputTwo == 2:
                print("Welcome Rohan")
                print("where you want to enter?")
                print("1. Diet ")
                print("2. Execerise")

                print("\n")

                user_inputThree = input(f"Enter your choice:")
                if user_inputThree == '1':
                    print("Diet Section")

                    while(True):
                        userMeal = input("Enter the Meal you have taken:")
                        if userMeal == 'q' or userMeal == 'Q':
                            print("Successfully inserted")
                            break
                        elif userMeal != 'q' or userMeal != 'Q':
                            f = open("Rohan_Diet.txt","a")
                            f.write(f"\n [{getdate()}]: {userMeal}")
                            f.close()
                            continue               

                elif user_inputThree == '2':
                    print("Excerise Section")

                    while(True):
                        userExercise = input("Enter the Exercise that you have Done:")
                        if userExercise == 'q' or userExercise == 'Q':
                            print("Successfully inserted")
                            break
                        elif userExercise != 'q' or userExercise != 'Q':
                            f = open("Rohan_Exercise.txt","a")
                            f.write(f"\n [{getdate()}]: {userExercise}")
                            f.close()
                            continue

                elif user_inputThree != '1' and user_inputThree != '2':
                    print("Wrong Input")


            elif user_inputTwo == 3:
                print("Welcome Hammad")
                print("Where you want to enter?")
                print("1. Diet ")
                print("2. Execerise")

                print("\n")

                user_inputThree = input(f"Enter your choice:")
                if user_inputThree == '1':
                    print("Diet Section")

                    while(True):
                        userMeal = input("Enter the Meal you have taken:")

                        if userMeal == 'q' or userMeal == 'Q':
                            print("Successfully inserted")
                            break
                        elif userMeal != 'q' or userMeal != 'Q':
                            f = open("Hammad_Diet.txt","a")
                            f.write(f"\n [{getdate()}]: {userMeal}")
                            f.close()
                            continue


                elif user_inputThree == '2':
                    print("Excerise Section")

                    while(True):
                        userExercise = input("Enter the Exercise that you have Done:")
                        if userExercise == 'q' or userExercise == 'Q':
                            print("Successfully inserted")
                            break
                        elif userExercise != 'q' or userExercise != 'Q':
                            f = open("Hammad_Exercise.txt","a")
                            f.write(f"\n [{getdate()}]: {userExercise}")
                            f.close()
                            continue

                elif user_inputThree != '1' and user_inputThree != '2':
                    print("Wrong Input")
            else:
                print("Oops wrong input :( )")

        elif user_inputOne == 2:
            Retrieve()

        elif user_inputOne == 3:
            print("Thanks for accessing this software....")
            break
            
        continue

manageSystem()
    
        
    
```
